Because of the small size of the dataframe I needed, and the inconsistent structure of the source files, I imputed the 'pms' values manually

```python
pms_data = pd.DataFrame({
    'month': ["Jan 2025", "Feb 2025", "Mar 2025", "Apr 2025", "May 2025", "Jun 2025", "Jul 2025", "Aug 2025", "Sep 2025", "Oct 2025", "Nov 2025", "Dec 2025", "Jan 2026"],
    'avg_pms_price': ["1258.34", "1245.80", "1261.65", "1239.33", "1027.76", "1037.66", "1024.99", "988.25", "970.59", "1052.31", "1061.35", "1048.63", "1034.76"]
})
pms_data['month'] = pd.to_datetime(
    pms_data['month'], format='%b %Y'
).dt.to_period('M')

pms_data['avg_pms_price'] = pd.to_numeric(
    pms_data['avg_pms_price']
)

```
